# Titanic Survival Classifier

```python
import pandas as pd

from src.titanic import TitanicDataset, TitanicClassifier
```

### Step 1: Data Preprocessing & Cleansing

In this step, we preprocess and cleanse the raw data from the Titanic Dataset which was provided from Kaggle.

We can import the data using Pandas `pd.read_csv` method and then use the class `TitanicDataset` to preprocess the data.

```python
raw_train_df = pd.read_csv('data/train.csv')
raw_test_df = pd.read_csv('data/test.csv')
```

```python
dataset = TitanicDataset()
train_df, test_df = dataset.preprocess(raw_train_df, raw_test_df)
```

### Step 2: Data Visualization

Use the `TitanicDataset` methods for visualization, such as `heatmap` and `visualize`

```python
dataset.heatmap(train_df)
```

```python
dataset.visualize(train_df)
```

### Step 3: Training / fitting the model

We fit a Random Forest Classifier Model using the `TitanicClassifier` class by simply plugging in the `train_df` that was preprocessed by the `TitanicDataset`.

```python
X_train = train_df.drop(columns=['Survived'])
y_train = train_df['Survived']

classifier = TitanicClassifier()
classifier.train(X_train, y_train)
```

### Step 4: Testing the model

Test the model by using the `test_df` which is basically from the `data/test.csv`

```python
X_test = test_df
predictions = classifier.predict(X_test)
predictions
```

We can also output in the form of a `.csv` file, which is useful when you are going to the submit Kaggle.

```python
classifier.output_csv(
    raw_test_df['PassengerId'],
    predictions,
    'results/submission.csv'
)
```
